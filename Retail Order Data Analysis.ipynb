**Questions**

```python
#Find top 10 highest revenue generating products

mycursor.execute('''SELECT product_id, SUM(sale_price) AS total_revenue FROM REATAILORDER1.RETAILORDER GROUP BY product_id ORDER BY total_revenue DESC LIMIT 10;''')
table = mycursor.fetchall()
answer_1 = pd.DataFrame(table, columns=['product_id', 'total_revenue'])
display(answer_1)
```

```python
#Find the top 5 cities with the highest profit margins

mycursor.execute('''SELECT city, SUM(proft) AS total_profit FROM REATAILORDER1.RETAILORDER GROUP BY city ORDER BY total_profit DESC LIMIT 5;''')
table = mycursor.fetchall()
answer_2 = pd.DataFrame(table, columns=['city', 'total_profit'])
display(answer_2)
```

```python
#Calculate the total discount given for each category

mycursor.execute('''SELECT category, SUM(discount) AS total_discount FROM REATAILORDER1.RETAILORDER GROUP BY category;''')
table = mycursor.fetchall()
answer_3 = pd.DataFrame(table, columns=['category', 'total_discount'])
display(answer_3)
```

```python
#Find the average sale price per product category

mycursor.execute('''SELECT category, AVG(sale_price) AS average_sale_price FROM REATAILORDER1.RETAILORDER GROUP BY category;''')

table = mycursor.fetchall()
answer_4 = pd.DataFrame(table, columns=['category', 'average_sale_price'])
display(answer_4)
```

```python
#Find the region with the highest average sale price

mycursor.execute('''SELECT region, AVG(sale_price) AS average_sale_price FROM REATAILORDER1.RETAILORDER GROUP BY region ORDER BY average_sale_price DESC LIMIT 1;''')

table = mycursor.fetchall()
answer_5 = pd.DataFrame(table, columns=['region', 'average_sale_price'])
display(answer_5)
```

```python
#Find the total profit per category

mycursor.execute('''SELECT category, SUM(proft) AS total_profit FROM REATAILORDER1.RETAILORDER GROUP BY category;''')

table = mycursor.fetchall()
answer_6 = pd.DataFrame(table, columns=['category', 'total_profit'])
display(answer_6)
```

```python
#Identify the top 3 segments with the highest quantity of orders

mycursor.execute('''SELECT segment, SUM(quantity) AS total_quantity FROM REATAILORDER1.RETAILORDER GROUP BY segment ORDER BY total_quantity DESC LIMIT 3;''')

table = mycursor.fetchall()
answer_7 = pd.DataFrame(table, columns=['segment', 'total_quantity'])
display(answer_7)
```

```python
#Determine the average discount percentage given per region

import pandas as pd
mycursor.execute('''SELECT region, AVG(discount_persent) AS average_discount_percentage FROM REATAILORDER1.RETAILORDER GROUP BY region;''')

table = mycursor.fetchall()
answer_8 = pd.DataFrame(table, columns=['region', 'average_discount_percentage'])
display(answer_8)
```

```python
#Find the product category with the highest total profit

mycursor.execute('''SELECT category, SUM(proft) AS total_profit FROM REATAILORDER1.RETAILORDER GROUP BY category ORDER BY total_profit DESC LIMIT 1;''')

table = mycursor.fetchall()
answer_9 = pd.DataFrame(table, columns=['category', 'total_profit'])
display(answer_9)
```

```python
#Calculate the total revenue generated per year

mycursor.execute('''SELECT YEAR(order_date) AS year, SUM(sale_price) AS total_revenue FROM REATAILORDER1.RETAILORDER GROUP BY year ORDER BY year;''')

table = mycursor.fetchall()
answer_10 = pd.DataFrame(table, columns=['year', 'total_revenue'])
display(answer_10)
```

```python
#Find the total sales for each product category.

mycursor.execute('''SELECT category, SUM(sale_price) AS total_sales FROM REATAILORDER1.RETAILORDER GROUP BY category;''')
table = mycursor.fetchall()
answer_11 = pd.DataFrame(table, columns=['category', 'total_sales'])
display(answer_11)
```

```python
#Find the average profit margin for each ship mode.

mycursor.execute(''' SELECT ship_mode, AVG(proft) AS average_profit_margin FROM REATAILORDER1.RETAILORDER GROUP BY ship_mode;''')

table = mycursor.fetchall()
answer_12 = pd.DataFrame(table, columns=['ship_mode', 'average_profit_margin'])
display(answer_12)
```

```python
# Find the top 5 customers who made purchases in the 'Furniture' category.

mycursor.execute('''SELECT order_id, COUNT(*) AS purchase_count FROM REATAILORDER1.RETAILORDER WHERE category = 'Furniture' GROUP BY order_id ORDER BY purchase_count DESC LIMIT 5;''')

table = mycursor.fetchall()
answer_13 = pd.DataFrame(table, columns=['order_id', 'purchase_count'])
display(answer_13)
```

```python
#Find the total revenue generated in each region for the year 2017.

mycursor.execute('''SELECT region, SUM(sale_price) AS total_revenue FROM REATAILORDER1.RETAILORDER WHERE YEAR(order_date) = 2017 GROUP BY region;''')

table = mycursor.fetchall()
answer_14 = pd.DataFrame(table, columns=['region', 'total_revenue'])
display(answer_14)
```

```python
#Find the products that have been ordered more than 10 times.

mycursor.execute('''SELECT product_id, COUNT(order_id) AS order_count FROM REATAILORDER1.RETAILORDER GROUP BY product_id HAVING order_count > 10;''')

table = mycursor.fetchall()
answer_15 = pd.DataFrame(table, columns=['product_id', 'order_count'])
display(answer_15)
```

```python
#Find the average quantity of products ordered per segment.

mycursor.execute('''SELECT segment, AVG(quantity) AS average_quantity FROM REATAILORDER1.RETAILORDER GROUP BY segment;''')

table = mycursor.fetchall()
answer_16 = pd.DataFrame(table, columns=['segment', 'average_quantity'])
display(answer_16)
```

```python
#Find the total profit generated by each product sub-category.

mycursor.execute('''SELECT sub_category, SUM(proft) AS total_profit FROM REATAILORDER1.RETAILORDER GROUP BY sub_category;''')

table = mycursor.fetchall()
answer_17 = pd.DataFrame(table, columns=['sub_category', 'total_profit'])
display(answer_17)
```

```python
#Find the top 3 states with the highest total sales.

mycursor.execute('''SELECT state, SUM(sale_price) AS total_sales FROM REATAILORDER1.RETAILORDER GROUP BY state ORDER BY total_sales DESC LIMIT 3;''')

table = mycursor.fetchall()
answer_18 = pd.DataFrame(table, columns=['state', 'total_sales'])
display(answer_18)
```

```python
#Find the total discount amount given for each product category.

mycursor.execute('''SELECT category, SUM(discount) AS total_discount FROM REATAILORDER1.RETAILORDER GROUP BY category;''')

table = mycursor.fetchall()
answer_19 = pd.DataFrame(table, columns=['category', 'total_discount'])
display(answer_19)
```

```python

#Find the average list price of products within each category and sub-category.

mycursor.execute(''' SELECT category, sub_category, AVG(list_price) AS average_list_price FROM REATAILORDER1.RETAILORDER GROUP BY category, sub_category;''')

table = mycursor.fetchall()
answer_20 = pd.DataFrame(table, columns=['category', 'sub_category', 'average_list_price'])
display(answer_20)
```
